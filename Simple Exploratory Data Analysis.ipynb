```python
import numpy as np
# your code goes here

Q1 = df.quantile(0.25)

Q3 = df.quantile(0.75)

IQR = Q3 - Q1

print(IQR)

print(df['ConvertedComp'].quantile(0.10))

print(df['ConvertedComp'].quantile(0.90))


print(df['ConvertedComp'].skew())
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
df_out = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]

print(df_out.shape)
```
